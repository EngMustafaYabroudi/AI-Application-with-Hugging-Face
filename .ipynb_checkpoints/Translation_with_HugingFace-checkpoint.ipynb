```python


import torch
import torch.nn as nn
import transformers

# تعريف النموذج البسيط
class SimpleModel(nn.Module):
    def __init__(self):
        super(SimpleModel, self).__init__()
        self.fc = nn.Linear(10, 2)  # نموذج بسيط بطبقة خطية

    def forward(self, x):
        return self.fc(x)

```

```python
# إنشاء النموذج
model = SimpleModel()
print(model)
```

```python
# حفظ النموذج إلى ملف
model_path = '/content/simple_model.pth'
torch.save(model.state_dict(), model_path)

# تحميل النموذج من الملف
model.load_state_dict(torch.load(model_path))
```
